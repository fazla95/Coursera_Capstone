## Demo: How to work with database


Download database file.

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp 
FROM master
"""

df_cc = pd.read_sql_query(QUERY,conn)
df_cc.hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age 
FROM master
"""

df_age = pd.read_sql_query(QUERY,conn)
df_age.boxplot()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs  
FROM master
"""

df_aw = pd.read_sql_query(QUERY,conn)
df_aw.plot.scatter(x = 'Age', y = 'WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs  
FROM master
"""

df_awc = pd.read_sql_query(QUERY,conn)
df_awc.plot.scatter(x = 'WorkWeekHrs', y = 'CodeRevHrs', s = 'Age')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as Total
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Total DESC
LIMIT 5
"""

df_ddny = pd.read_sql_query(QUERY,conn)
df_ddny['Total'].plot(kind='pie', labels = df_ddny['DatabaseDesireNextYear'], ylabel = '', title = 'DatabaseDesireNextYear', autopct='%1.1f%%')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs  
FROM master
WHERE AGE BETWEEN 30 and 35
"""

df_awc = pd.read_sql_query(QUERY,conn)
Age = [30, 31, 32, 33, 34, 35]
median_wwh, median_crh = [], []
for age in Age :
    median_wwh.append(df_awc[df_awc['Age']==age]['WorkWeekHrs'].median())
    median_crh.append(df_awc[df_awc['Age']==age]['CodeRevHrs'].median())
df = pd.DataFrame({'Median_WWH' : median_wwh, 'Median_CRH' : median_crh}, index = Age)
df.plot.bar(stacked=True)
median_wwh, median_crh
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE AGE BETWEEN 45 and 60
"""

df_acc = pd.read_sql_query(QUERY,conn)
Age = range(45,61)
median_cc = []
for age in Age :
    median_cc.append(df_acc[df_acc['Age']==age]['ConvertedComp'].median())
df = pd.DataFrame({'Median_CC' : median_cc}, index = Age)   
df.plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(MainBranch) as Total
FROM master
GROUP BY MainBranch
"""

df_mb = pd.read_sql_query(QUERY,conn)
df_mb.plot.barh()
```
